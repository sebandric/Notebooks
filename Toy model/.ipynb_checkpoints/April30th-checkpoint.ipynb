## Another model

### Set up environment

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.collections import LineCollection
from scipy.integrate import odeint
from scipy.interpolate import UnivariateSpline
from ddeint import ddeint
```

### Stim functions

Used to generate the input pulses I(t)

```python
# Input functions
def rect(T):
    """
    create a centered rectangular pulse of width $T
    """
    return lambda t: (-T/2 <= t) & (t < T/2)

def pulse_train(t, at, shape):
    """
    create a train of pulses over $t at times $at and shape $shape
    """

    return np.sum(shape(t - at[:,np.newaxis]), axis=0)
```

### System

System of diff. eqs. for the model. 
* d_theta0 is optional, only included if the input is modeled as an oscillator rather than pulses.
* d_theta1 is the input layer (sensory area, low freq. oscillator)
* d_theta2 is the feedback layer (motor area, high freq. bursting oscillator)

```python
def nonuniform(X, t, d, omegas, gains, nonunifs, I):
    # Set initial conditions and params 
    theta0,theta1,theta2 = X(t)
    theta0d,theta1d,theta2d, = X(t-d)
    
    w0,w1,w2 = omegas
    k1,k2 = gains
    a0,a1,a2 = nonunifs
    
    # Input signal (Optional, currently not interacting with system)
    d_theta0 = w0 + a0*np.sin(theta0)
    
    # First layer
    d_theta1 = w1 + a1*(1+np.sin(theta2))*np.sin(theta1) + k1*I(t)*np.cos(theta1)
    
    # Second layer  
    d_theta2 = w2 + a2*np.sin(theta2) + k2*np.sin(theta1-theta1d) 
     
    return np.array([d_theta0, d_theta1, d_theta2])
```

### Parameters

* thetas: intial conditions of each oscillator
* d: delay at which the top layer compares the phase difference (theta(t) - (theta(t-d))
* frequencies: oscillator frequencies in Hz
* w: oscillator angular frequecies
* k: gain parameter for each oscillator 
* a: nonuniformities

```python
# Initial Conditions
thetas = lambda tt: np.array([1, 2, 3])

# Delay used for layer 2
d = 5

# Parameter values
frequencies = frequency0, frequency1, frequency2 = 3, 4, 15
w0 = (frequency0*2*np.pi/1000)
w1 = (frequency1*2*np.pi/1000)
w2 = (frequency2*2*np.pi/1000)
k1,k2 = 1, 0.5
a0,a1,a2 = 0.2*w0, 0.2*w1, 0.2*w2

omegas = [w0,w1,w2]
gains = [k1,k2]
nonunifs = [a0,a1,a2]
```

### Time series and stim set-up

```python
# time series
tstart = 0.0
tend = 5000
tnum = 10000
tstep = (tend-tstart) / tnum
t = np.linspace(tstart, tend, tnum)

# start time, freq., and period of stim
stim_start = 1000
stim_frequency = 3
stim_period = (1/stim_frequency)*1000

# stim index to be perturbed and magnitude of pertrubation (set p_mag to zero for isochronous stimulus)
p_index = 5
p_mag = 40

# set-up stim
onsets = np.arange(stim_start,tend,stim_period)
onsets[p_index] = onsets[p_index] + p_mag
sig = pulse_train(
    t,             # time domain
    at=onsets,     # times of pulses
    shape=rect(1)  # shape of pulse
)

# interpolate stimulus for DDE
I = UnivariateSpline(t, sig)
I.set_smoothing_factor(0)
```

### Solve

```python
# Combine args
args = (d,omegas,gains,nonunifs,I)

# Solve
sol = ddeint(nonuniform, thetas, t, fargs=args)
S = sol[:,0]
X = sol[:,1]
Y = sol[:,2]
```

```python
figsize = (16,3)

# Plot stimulus 
plt.figure(figsize=figsize)
plt.plot(t,I(t),'k')
perturb_ts = t[t > (onsets[p_index]-50)]
perturb_ts = perturb_ts[perturb_ts<(onsets[p_index]+50)]
plt.plot(perturb_ts,I(perturb_ts),'g')

# Plot sinusoidal stimulus (NOT IN SYSTEM)
#plt.figure(figsize=figsize)
#plt.vlines(perturb,min(np.sin(S)),max(np.sin(S)),linestyles='dotted')
#plt.plot(t,np.sin(S),'k')

# Plot Layer 1
plt.figure(figsize = figsize)
plt.vlines(onsets,min(np.sin(X)),max(np.sin(X)),linestyles='dotted')
plt.plot(t,np.sin(X),'r')

# Plot layer 2
plt.figure(figsize = figsize)
plt.vlines(onsets,min(np.sin(Y)),max(np.sin(Y)),linestyles='dotted')
plt.plot(t,np.sin(Y),'m')
```
